```python
class Student:

    def __init__(self, name, scores):
        self.name: str = name
        self.scores: list = scores

    def calculate_average(self):
        return round(sum(self.scores) / len(self.scores), 2)

    def is_passing(self):
        for score in self.scores:
            if score < 40:
                return "Fail"
        return "The student is passing all subjects"


class Performance_Tracker:
    def __init__(self):
        self.students = {}

    def add_student(self, student):
        self.students[student.name] = student

    def calculate_class_average(self):
        total_score = 0
        total_students = len(self.students)
        for student in self.students.values():
            total_score += student.calculate_average()
        if total_students > 0:
            return round(total_score / total_students, 2)
        return 0

    def display_student_performance(self):
        if not self.students:
            print("No student data available.")
            return
        for name, student in self.students.items():
            average_score = student.calculate_average()
            passing_status = student.is_passing()
            print(f"Student: {name}")
            print(f"  Average Score: {average_score}")
            print(f"  Passing Status: {passing_status}")
            print("-" * 20)


def input_student_data():
    students_data = {}
    while True:
        student_name = input("Enter student's name (or type 'done' to finish): ")
        if student_name.lower() == 'done':
            break

        scores = []
        for subject in ["math", "science", "english"]:
            while True:
                try:
                  score = int(input(f"Enter {subject} score for {student_name}: "))
                  if 0 <= score <= 100:
                      scores.append(score)
                      break
                  else:
                      print("Score must be between 0 and 100.")
                except ValueError:
                    print("Invalid input. Please enter a valid integer score.")
        students_data[student_name] = scores
    return students_data
```

```python
student1 = Student(name="Waqar", scores=[50, 80, 70, 17, 69])
student2 = Student(name="Ali", scores=[90, 95, 85, 75, 80])

tracker = Performance_Tracker()
tracker.add_student(student1)
tracker.add_student(student2)

print(f"Class Average: {tracker.calculate_class_average()}")
tracker.display_student_performance()
```

```python
students_data = input_student_data()

performance_tracker = Performance_Tracker()
for name, scores in students_data.items():
  student = Student(name, scores)
  performance_tracker.add_student(student)


performance_tracker.display_student_performance()
class_average = performance_tracker.calculate_class_average()
print(f"Class Average: {class_average}")
```
